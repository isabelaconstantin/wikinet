```python
import pandas as pd

# Download the dataset in the directory /dataset/wikispeedia_paths-and-graph
# source: https://snap.stanford.edu/data/wikispeedia/wikispeedia_paths-and-graph.tar.gz

articles = pd.read_csv('dataset/wikispeedia_paths-and-graph/articles.tsv', sep='\t', names=['article'], skiprows=12)
links = pd.read_csv('dataset/wikispeedia_paths-and-graph/links.tsv', sep='\t', names=['article', 'link'], skiprows=12)

print(articles)
links
```

```python
import networkx as nx

G = nx.Graph()
G.add_nodes_from(articles['article'])

G.number_of_nodes()
```

```python
G.add_edges_from(links.itertuples(index=False))
G.number_of_edges()
```
